```python
# Avg daily return Pre & Post Covid

avg_daily_return_sp500 = daily_returns_pre_covid_refined.mean()['S&P500']


```

```python
# Compute daily volatility - change it 
std_dev_daily_return_precovid_sp500 = daily_returns_pre_covid_refined.std()['S&P500']

print(f"Daily Volatility of S&P 500 stock is: {std_dev_daily_return_sp500}")
```

```python
# Save the last day's closing price - change it 
sp500_last_price = pre_covid_df['S&P500'][-1]

print(f"Last Day's Closing Price of S&P 500 stock is: {sp500_last_price }")
```

```python
# Setup the Monte Carlo Parameters
number_simulations = 1000
number_records = 252 * 20


# Initialize empty DataFrame to hold simulated prices for each simulation
monte_carlo = pd.DataFrame()
sp500_cumulative_returns = pd.DataFrame()
```

```python
# Run the Monte Carlo Simulation

# Run the simulation of projecting stock prices for the next trading year, 600` times
for n in range(number_simulations):

    # Initialize the simulated prices list with the last closing price of `SPY` and `AGG`
    simulated_sp500_prices = [sp500_last_price]
    
    
    # Simulate the returns for 252 * 20 days
    for i in range(number_records):
        
        # Calculate the simulated price using the last price within the list
        simulated_sp500_price = simulated_sp500_prices[-1] * (1 + np.random.normal(avg_daily_return_sp500, std_dev_daily_return_precovid_sp500))

        
        # Append the simulated price to the list
        simulated_sp500_prices.append(simulated_sp500_price)
    
    # Append the simulated prices of each simulation to DataFrame
    monte_carlo= pd.Series(simulated_sp500_prices)
    
    
    # Calculate the daily returns of simulated prices
    simulated_daily_returns_pre_covid = monte_carlo.pct_change()
    
#     # Set the portfolio weights (60% SPY; 40% AGG)
#     weights = [0.60, 0.40]

#     # Use the `dot` function with the weights to multiply weights with each column's simulated daily returns
#     portfolio_daily_returns = simulated_daily_returns.dot(weights)
    
    # Calculate the normalized, cumulative return series
    sp500_cumulative_returns[f"Simulation {n+1}"] = (1 + simulated_daily_returns_pre_covid.fillna(0)).cumprod()

sp500_cumulative_returns.head()
```

```python
# To determine forecasted stock price, convert cumulative returns back to price by multiplying it by last price of the stock

forecasted_price_sp500 = sp500_cumulative_returns*sp500_last_price

forecasted_price_sp500.head()
```

```python
# Calculate average price of all simulations for the time period
avg_forecasted_price_sp500 = forecasted_price_sp500.mean(axis =1)
avg_forecasted_price_sp500.count()
avg_forecasted_price_sp500.head()
```

```python
# Select first 95 records and convert to dataframe for comparison with actual prices
avg_price_sp500_comparison = avg_forecasted_price_sp500.iloc[0:95].to_frame()
avg_price_sp500_comparison.columns = ['forecasted_price']
avg_price_sp500_comparison.count()
```

```python
avg_price_sp500_comparison.head()
```

```python
# Setup the Monte Carlo Parameters
number_simulations = 1000
number_records = 252 * 20

# Initialize empty DataFrame to hold simulated prices for each simulation
monte_carlo = pd.DataFrame()
ftse100_cumulative_returns = pd.DataFrame()
```

```python
# Run the Monte Carlo Simulation

# Run the simulation of projecting stock prices for the next trading year, 600` times
for n in range(number_simulations):

    # Initialize the simulated prices list with the last closing price of `SPY` and `AGG`
    simulated_ftse100_prices = [ftse100_last_price]
    
    
    # Simulate the returns for 252 * 20 days
    for i in range(number_records):
        
        # Calculate the simulated price using the last price within the list
        simulated_ftse100_price = simulated_ftse100_prices[-1] * (1 + np.random.normal(avg_daily_return_ftse100, std_dev_daily_return_ftse100))

        
        # Append the simulated price to the list
        simulated_ftse100_prices.append(simulated_ftse100_price)
    
    # Append the simulated prices of each simulation to DataFrame
    monte_carlo= pd.Series(simulated_ftse100_prices)
    
    
    # Calculate the daily returns of simulated prices
    simulated_daily_returns_pre_covid = monte_carlo.pct_change()
    
#     # Set the portfolio weights (60% SPY; 40% AGG)
#     weights = [0.60, 0.40]

#     # Use the `dot` function with the weights to multiply weights with each column's simulated daily returns
#     portfolio_daily_returns = simulated_daily_returns.dot(weights)
    
    # Calculate the normalized, cumulative return series
    ftse100_cumulative_returns[f"Simulation {n+1}"] = (1 + simulated_daily_returns_pre_covid.fillna(0)).cumprod()

ftse100_cumulative_returns.head()
```

```python
# Convert cumulative returns back to price by multiplying it by last price of the stock

forecasted_price_ftse100 = ftse100_cumulative_returns*ftse100_last_price

forecasted_price_ftse100.head()
```

```python
# Calculate average price of all simulations for the time period
avg_forecasted_price_ftse100 = forecasted_price_ftse100.mean(axis =1)
avg_forecasted_price_ftse100.count()
#avg_price_sha.head()
```
